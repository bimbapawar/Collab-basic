Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.



Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]


Constraints:

2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109
Only one valid answer exists.

## ðŸ“Œ Problem Reminder (Two Sum)

Given:

```python
nums   = [2, 7, 11, 15]
target = 9
```

We must return the **indices** of two numbers whose sum equals the target.

---

## ðŸ§  Code to Explain

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        di = {}
        for i, v in enumerate(nums):
            needed = target - v
            if needed in di:
                return (di[needed], i)
            di[v] = i
```

---

### ðŸ”¹ `return (di[needed], i)`

* Returns:

  * Index of the earlier number
  * Index of the current number

---

### ðŸ”¹ `di[v] = i`

* Stores the current number and its index
* So it can be used later

## ðŸ§  One-Line Beginner Explanation

> We store numbers we have seen in a dictionary, and for each new number we check if its required pair already exists.







````python
nums = [2,7,11,15]
target = 9
di = {}

for i, v in enumerate(nums):
    needed = target - v
    print(f"i={i}, v={v}, needed={needed}, di={di}")
    if needed in di:
        print("Answer:", di[needed], i)
        break
    di[v] = i
````
